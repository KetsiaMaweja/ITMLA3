```python
import os
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import precision_score, recall_score, f1_score
from skimage.feature import hog
from sklearn.preprocessing import MinMaxScaler
 
def load_images_and_labels(folder_path, target_size=(64, 64), count=None):
    images = []
    labels = []
 
    for root, dirs, files in os.walk(folder_path):
        for filename in files:
            image_path = os.path.join(root, filename)
            # Extract the label from the immediate parent folder
            label = os.path.basename(os.path.dirname(image_path))
            image = cv2.imread(image_path)
            if image is not None:
                image = cv2.resize(image, target_size)
                images.append(image)
                labels.append(label)
 
    return images, labels
 
def convert_to_gray(images):     
    gray_images = []     
    for image in images:        
        # Convert to grayscale        
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)         
        gray_images.append(gray_image)     
    return gray_images
 
def image_enhancement(images, labels):
    enhanced_images = []
    gray_imgs = convert_to_gray(images)
    for image in gray_imgs:
        # Apply histogram equalization
        enhanced_image = cv2.equalizeHist(image)
        enhanced_images.append(enhanced_image)
    return enhanced_images
directory = r"C:\Users\Admin\Downloads\itmla3_project_dataset\CONCRETE CRACKS"
images, labels = load_images_and_labels(directory)
enhanced_images = image_enhancement(images, labels)
 
if enhanced_images:
    plt.figure(figsize=(8, 4))
    plt.imshow(enhanced_images[0], cmap='gray')
    plt.title('Equalized Image - ' + labels[0])
    plt.axis('off')
    plt.show()
 
    plt.figure(figsize=(8, 4))
    plt.imshow(enhanced_images[-1], cmap='gray')
    plt.title('Equalized Image - ' + labels[-1])
    plt.axis('off')
    plt.show()

def split_data(features, labels, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test
 
def feature_normalization(X):
    # Reshape images for MLP input
    X_flatten = [image.flatten() for image in X]
    # Normalize features using min-max scaling
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X_flatten)
    return X_scaled, scaler
 
def train_model(X_train, y_train):
    # Normalize features
    X_train_scaled, scaler = feature_normalization(X_train)
    # Train the model
    clf = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000)
    clf.fit(X_train_scaled, y_train)
    return clf, scaler
 
def test_model(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred
 
def calculate_metrics(y_test, y_pred):
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    print("Precision: {:.2f}".format(precision))
    print("Recall: {:.2f}".format(recall))
    print("F1-score: {:.2f}".format(f1))
 
 
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = split_data(enhanced_images, labels)
 
# Feature normalization
X_train_normalized, scaler = feature_normalization(X_train)
X_test_normalized = scaler.transform([image.flatten() for image in X_test])
 
# Train the model
model, _ = train_model(X_train_normalized, y_train)
 
# Test the model
y_pred = test_model(model, X_test_normalized)
 
# Calculate and print metrics
calculate_metrics(y_test, y_pred)
```
